```python
for factorx in np.arange(1,3.3125,0.0875):
    plt.figure(figsize=(5,5))

    side = 30
    factory = factorx

    plt.xlim([-side/2/factorx, side/2/factorx])
    plt.ylim([-side/2/factory, side/2/factory])
    plt.axis('off') 

    line = np.arange(-side/2, side/2+1)
    for number in np.arange(-side/2, side/2+2,2):
        ax = number*np.ones(len(line)) 
        plt.plot(ax, line, 'k-')

    for number in np.arange(-side/2, side/2+2,2):
        ay = number*np.ones(len(line))
        plt.plot(line, ay, 'k-')

    plt.scatter(0,0,s=240,c='red')
    for theta in np.arange(0, 2*np.pi, 0.1666666666666666666666*np.pi):
        for R in np.arange(2, 15, 2):
            x = R*np.cos(theta)
            y = R*np.sin(theta)
            s = 150 - 10*R
            plt.scatter(x,y,s,c='red')
            
#    plt.savefig('GIFInhIso %f.png' %factorx)
```
